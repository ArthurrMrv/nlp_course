## Exercise 1: Function Plotting and Analytical Determination of Critical Points

Determine analytically the coordinates and nature of the critical points of the following functions:

- $f_1(x,y) = x^2 + y^2 - 5x - 12y$
- $f_2(x,y) = x^3 - 12xy + 8y^3$
- $f_3(x,y) = (x-1)^2 + 10(x^2 - y^2)$

**Vocabulary Note:** An 'analytical' determination means that the calculation must be 'exact', without using numerical tools or formal computation.


```python
# Cell 1.1: Import necessary libraries
import numpy as np
import sympy as sym
import matplotlib.pyplot as plt
```

```python
# Cell 1.2: Define the functions
def f1(x, y):
    """f1(x,y) = x^2 + y^2 - 5x - 12y"""
    return x**2 + y**2 - 5*x - 12*y

def f2(x, y):
    """f2(x,y) = x^3 - 12xy + 8y^3"""
    return x**3 - 12*x*y + 8*y**3

def f3(x, y):
    """f3(x,y) = (x-1)^2 + 10(x^2 - y^2)"""
    return (x-1)**2 + 10*(x**2 - y**2)
```

```python
# Cell 3.3: Define the gradient of f1(x,y)
def dF1(x, y):
    """Gradient of f1(x,y) = x^2 + y^2 - 5x - 12y"""
    df_dx = 2*x - 5
    df_dy = 2*y - 12
    return [df_dx, df_dy]
```

```python
# Cell 3.4: Calculate gradient at a specific point
gradient = dF1(2.5, 6)
print(f"Gradient at (2.5, 6): {gradient}")
```

### Instructions 3.4

Complete the template for $d^2F_1$, the Hessian matrix of $f_1(x,y)$, and verify the nature of critical points.


```python
# Cell 3.5: Define the Hessian matrix of f1(x,y)
def d2F1(x, y):
    """Hessian matrix of f1(x,y) = x^2 + y^2 - 5x - 12y"""
    d2f_dx2 = 2
    d2f_dy2 = 2
    d2f_dxdy = 0
    return [[d2f_dx2, d2f_dxdy], [d2f_dxdy, d2f_dy2]]
```

```python
# Cell 3.6: Calculate eigenvalues of the Hessian matrix for f1 at (2.5, 6)
hessian_matrix = np.array(d2F1(2.5, 6))
eigenvalues = np.linalg.eigvals(hessian_matrix)
print(f"Eigenvalues of the Hessian: {eigenvalues}")
print(f"First eigenvalue: {eigenvalues[0]}")
print(f"Second eigenvalue: {eigenvalues[1]}")
```

```python
# Cell 3.7: Calculate eigenvalues of the Hessian matrix for f2 at (-2, -1)
# First, we need to compute the Hessian for f2 at this point
# For f2(x,y) = x^3 - 12xy + 8y^3:
# df/dx = 3x^2 - 12y
# df/dy = -12x + 24y^2
# d2f/dx2 = 6x
# d2f/dy2 = 48y
# d2f/dxdy = -12

x_val, y_val = -2, -1
d2f_dx2 = 6 * x_val
d2f_dy2 = 48 * y_val
d2f_dxdy = -12

hessian_matrix = np.array([[d2f_dx2, d2f_dxdy], [d2f_dxdy, d2f_dy2]])
eigenvalues = np.linalg.eigvals(hessian_matrix)
print(f"Eigenvalues of the Hessian: {eigenvalues}")
print(f"First eigenvalue: {eigenvalues[0]}")
print(f"Second eigenvalue: {eigenvalues[1]}")
```
